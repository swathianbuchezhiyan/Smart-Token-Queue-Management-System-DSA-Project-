```python
import time

queue = []
token_counter = 1


def add_customer(name, priority):
    global token_counter
    queue.append((token_counter, name, priority))
    print("Token", token_counter, "assigned to", name)
    token_counter += 1


def show_queue():
    if len(queue) == 0:
        print("Queue is empty")
        return

    print("Current Queue")
    for q in queue:
        status = "Emergency" if q[2] == 0 else "Normal"
        print("Token", q[0], "-", q[1], "-", status)


def serve_customer():
    if len(queue) == 0:
        print("Queue is empty")
        return

    queue.sort(key=lambda x: x[2])

    customer = queue.pop(0)

    status = "Emergency" if customer[2] == 0 else "Normal"
    print("Serving Token", customer[0], "-", customer[1], "-", status)

    if customer[2] == 0:
        time.sleep(1)
    else:
        time.sleep(2)


add_customer("Patient A", 1)
add_customer("Patient B", 0)
add_customer("Patient C", 1)
add_customer("Patient D", 0)

show_queue()

print("Serving Start")
serve_customer()
serve_customer()

show_queue()
```

```python
while True:
    print("\n1 Add Patient")
    print("2 Show Queue")
    print("3 Serve Patient")
    print("4 Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter patient name: ")
        priority = int(input("Enter priority (0 Emergency, 1 Normal): "))
        add_customer(name, priority)

    elif choice == "2":
        show_queue()

    elif choice == "3":
        serve_customer()

    elif choice == "4":
        print("System Closed")
        break

    else:
        print("Invalid Choice")
```
